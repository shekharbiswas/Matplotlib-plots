```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib._color_data as mcd
```

```python
df = pd.read_csv('columns_bar.csv')
df.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
df[]
```

```python
ax = df.plot(kind='bar', stacked=True, x = 'col0',figsize=(12, 12), fontsize=12, legend = False, color = mcd.CSS4_COLORS)

ax.set_xlabel("Country", fontsize = 18)
ax.set_ylabel("Observations", fontsize = 18)
ax.figure.savefig('barplot.png')
```

```python
ax = df.plot(kind='bar', stacked=True, x = 'col0',figsize=(12, 12), fontsize=12, legend = True, color = mcd.CSS4_COLORS)

handles,labels = ax.get_legend_handles_labels()
plt.legend(fontsize='xx-large', ncol=5,handleheight=2.4, labelspacing=0.05)
fig_legend = plt.figure(figsize=(30,15))
axi = fig_legend.add_subplot(111)            
fig_legend.legend(handles, labels, loc='center', fontsize='xx-large', ncol=5,handleheight=2.4, labelspacing=0.05)
axi.xaxis.set_visible(False)
axi.yaxis.set_visible(False)
fig_legend.canvas.draw()
fig_legend.show()
fig_legend.savefig('legend.png')
```
